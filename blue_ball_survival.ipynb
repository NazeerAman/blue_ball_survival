# Problem statement:
## Given two bags, one containing 21 blue and 23 red balls, and the other containing 22 red balls, we repeatedly draw two balls from the first bag. If the drawn balls are of different colors, the blue ball is returned; otherwise, both are removed, and a red ball from the second bag is added. This process continues until only one ball remains. What is the color of the final ball?

```python
import random

blue_ball="ğŸ”µ"
red_ball="ğŸ”´"
num_iteration=1

bag_with_red_blue_balls=[blue_ball]*21+[red_ball]*23
bag_with_red_balls=[red_ball]*22
random.shuffle(bag_with_red_blue_balls)
print("balls in bag1: ",bag_with_red_blue_balls)
print(f"Total number of balls in bag_with_red_blue_balls(bag-1): {len(bag_with_red_blue_balls)}"),
print(f"Total number of balls in bag_with_red_ball(bag-2): {len(bag_with_red_balls)}")
print("="*150,"\n"*4)



while len(bag_with_red_blue_balls) > 1:
    print("="*150,"\n"*2)
    print(f"for iteration {num_iteration}")
    selected_balls = random.sample(bag_with_red_blue_balls, 2)
    print("selected balls =",selected_balls)
    
    if selected_balls[0] != selected_balls[1]:
        if selected_balls[0] == red_ball or selected_balls[1] == red_ball:
            bag_with_red_blue_balls.remove(red_ball)
            print(f"{red_ball} is removed")
            print(f"{red_ball} is added")
        if len(bag_with_red_blue_balls) != 1:
            
            print("bag_with_red_blue_balls",bag_with_red_blue_balls)
        else:
            print("bag_with_red_blue_ball",bag_with_red_blue_balls)
    else:
        for ball in selected_balls:
            bag_with_red_blue_balls.remove(ball)
            print(f"{ball} is removed")
        bag_with_red_blue_balls.append(bag_with_red_balls.pop())
        print(f"{red_ball} is added")
        if len(bag_with_red_blue_balls) != 1:
            
            print("bag_with_red_blue_balls",bag_with_red_blue_balls)
        else:
            print("bag_with_red_blue_ball",bag_with_red_blue_balls)
            
    num_iteration+=1

```

```python
print("ball left in the bag : " ,len(bag_with_red_blue_balls))
print(f"after {num_iteration} iterations, ball left in bag1 is : ",bag_with_red_blue_balls)
```
